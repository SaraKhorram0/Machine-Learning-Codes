### Linear Regression Sample Model

Reading data using pandas

```python
import pandas as pd
from matplotlib import pyplot as plt

data = pd.read_csv('Data/Advertising.csv')

import seaborn as sns
%matplotlib inline
```

Visualizing data using seaborn

```python
sns.pairplot(data, x_vars=['Radio', 'TV', 'Newspaper'], y_vars=['Sales'], height=7, aspect=0.7,kind='reg')
plt.show()
```

Linear regression model:
$$y=\beta_0+\beta_1x_1+\beta_2x_2+...+\beta_nx_n$$
in this case:
$$y=\beta_0 + \beta_1 \times Radio + \beta_1 \times TV + \beta_2 \times Newspaper$$
- $y$ is response
- $\beta_0$ is the intercept
- $\beta_i$ is the model coefficient for $i_{th}$ feature

```python
feature_names = ['Radio', 'TV', 'Newspaper']
X = data[feature_names]
y = data['Sales']
```

Evaluation procedure: Train/Test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(X_train, y_train)
```

```python
print(linreg.coef_)
print(linreg.intercept_)
```

```python
list(zip(feature_names, linreg.coef_))
```

Feature selection:

```python
from sklearn import metrics
import numpy as np
y_pred = linreg.predict(X_test)
print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
feature_names = ['Radio', 'TV']
X = data[feature_names]
y = data['Sales']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
linreg = LinearRegression()
linreg.fit(X_train, y_train)
y_pred = linreg.predict(X_test)
print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
